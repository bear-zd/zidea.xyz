```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from  sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=50,centers=2,random_state=6)
plt.scatter(X[:,0],X[:,1],c=y,s=30)
```

```python
from sklearn.datasets import load_wine
#获得x,y的上下限
def make_meshgrid(x,y,h=0.02):
    x_min,x_max=x.min()-1,x.max()+1
    y_min,y_max=y.min()-1,y.max()+1
    xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    return xx,yy

#等高线
def plot_contours(ax,clf,xx,yy,**params):
    Z= clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z=Z.reshape(xx.shape)
    out=ax.contourf(xx,yy,Z,**params)
    return out
```

```python
wine=load_wine()#导入自带的wine数据集
X=wine.data[:,:2] #为保证数据为两维，使用前两个特征
Y=wine.target

C=1.0 #SVM的正则化参数
models=(svm.SVC(kernel='rbf',gamma=0.7,C=C),svm.SVC(kernel='poly',degree=3,C=C))
models=(clf.fit(X,Y) for clf in models)
```

```python

titles=('SVC with RBF kernel','SVC with polynomial (degree 3) kernel')
fig,sub=plt.subplots(2,1)
plt.subplots_adjust(wspace=0.4,hspace=0.4)
X0,X1=X[:,0],X[:,1]
xx,yy=make_meshgrid(X0,X1)

for clf,title,ax in zip(models,titles,sub.flatten()):
    plot_contours(ax,clf,xx,yy,cmap=plt.cm.plasma,alpha=0.8)
    ax.scatter(X0,X1,c=Y,cmap=plt.cm.plasma,s=20,edgecolors='k')
    ax.set_xlim(xx.min(),xx.max())
    ax.set_ylim(yy.min(),yy.max())
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
plt.show()
```

```python
from sklearn.datasets import load_wine
#获得x,y的上下限
def make_meshgrid(x,y,h=0.02):
    x_min,x_max=x.min()-1,x.max()+1
    y_min,y_max=y.min()-1,y.max()+1
    xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    return xx,yy

#等高线
def plot_contours(ax,clf,xx,yy,**params):
    Z= clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z=Z.reshape(xx.shape)
    out=ax.contourf(xx,yy,Z,**params)
    return out

wine=load_wine()#导入自带的wine数据集
X=wine.data[:,:2] #为保证数据为两维，使用前两个特征
Y=wine.target

C=1.0 #SVM的正则化参数
models=(svm.SVC(kernel='linear',C=C),svm.SVC(C=C))
models=(clf.fit(X,Y) for clf in models)

titles=('SVC with Linear kernel','LinearSVC(linear kernel)')
fig,sub=plt.subplots(2,1)
plt.subplots_adjust(wspace=0.4,hspace=0.4)
X0,X1=X[:,0],X[:,1]
xx,yy=make_meshgrid(X0,X1)

for clf,title,ax in zip(models,titles,sub.flatten()):
    plot_contours(ax,clf,xx,yy,cmap=plt.cm.plasma,alpha=0.8)
    ax.scatter(X0,X1,c=Y,cmap=plt.cm.plasma,s=20,edgecolors='k')
    ax.set_xlim(xx.min(),xx.max())
    ax.set_ylim(yy.min(),yy.max())
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
plt.show()
```

```python
C=1.0
models=(svm.SVC(kernel='rbf',gamma=0.1,C=C),svm.SVC(kernel='rbf',gamma=1,C=C),svm.SVC(kernel='rbf',gamma=1000,C=C))
models=(clf.fit(X,Y) for clf in models)

titles = ('gamma=0.1','gamma=1','gamma=10')
fig,sub=plt.subplots(1,3,figsize=(10,3))
X0,X1=X[:,0],X[:,1]
xx,yy=make_meshgrid(X0,X1)
for clf,title,ax in zip(models,titles,sub.flatten()):
    plot_contours(ax,clf,xx,yy,cmap=plt.cm.plasma,s=20,edgecolors='k')
    ax.scatter(X0,X1,c=Y,cmap=plt.cm.plasma,s=20,edgecolors='k')
    ax.set_xlim(xx.min(),xx.max())
    ax.set_ylim(yy.min(),yy.max())
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
plt.show()
```
